```python
sns.set_style("white")
f, ax=plt.subplots(figsize=(8,7))
ax = sns.boxplot(data=df[df_numeric],orient="h",palette="Set1")
ax.xaxis.grid(False)
```

```python
from sklearn.preprocessing import MinMaxScaler
df[df_numeric]
df_normalization = MinMaxScaler().fit_transform(df[df_numeric])
```

```python
df_normalization=pd.DataFrame(df_normalization)
```

```python
df_normalization.columns= ['age', 'fw', 'edu_num', 'cg', 'cl', 'hpw']
```

```python
df_normalization
```

```python
sns.set_style("white")
f, ax=plt.subplots(figsize=(8,7))
ax = sns.boxplot(data=df_normalization,orient="h",palette="Set1")
ax.xaxis.grid(False)
```

```python
for column in df_numeric:
    df_normalization[[column]]
    lower=df_normalization[[column]].quantile(0.01)
    lower=round(lower,2)
    upper=df_normalization[[column]].quantile(0.995)
    upper=round(upper,2)
    df_normalization[[column]]=df_normalization[[column]].clip(lower,upper,axis=1)
    print(lower,upper)
sns.set_style("white")
f, ax=plt.subplots(figsize=(8,7))
ax = sns.boxplot(data=df_normalization,orient="h",palette="Set1")
ax.xaxis.grid(False)    
```
